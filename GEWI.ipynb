```python
## ‚úÖ Bloco 3 ‚Äî Valida√ß√£o Sem√¢ntica com BERT + Similaridade Cosine

import sys
!{sys.executable} -m pip install pandas numpy tqdm sentence-transformers --quiet

import os
import pandas as pd
import numpy as np
from tqdm import tqdm
from sentence_transformers import SentenceTransformer, util

# Caminhos
input_base = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/"
output_base = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/validados/"
os.makedirs(output_base, exist_ok=True)

# ‚öôÔ∏è Carrega modelo BERT otimizado para portugu√™s
print("üîÅ Carregando modelo BERT...", flush=True)
modelo_bert = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')

# ‚ö° Frases-√¢ncora que representam e-WOM real
frases_ancora = [
    "comprei um disjuntor",
    "estou trocando a fia√ß√£o el√©trica da casa",
    "vou instalar novas tomadas e interruptores",
    "contratei um eletricista para refazer a instala√ß√£o",
    "preciso de um novo quadro de luz",
    "o fio queimou e tive que trocar",
    "tive curto-circuito em casa",
    "preciso de cabo el√©trico novo",
    "o eletricista recomendou trocar a caixa de disjuntores",
    "estou reformando a parte el√©trica do apartamento"
]
print("üìê Gerando embeddings das frases-√¢ncora...", flush=True)
embeddings_ancora = modelo_bert.encode(frases_ancora, convert_to_tensor=True)

# Per√≠odo de an√°lise: Jan/2010 at√© Jul/2023
anos_meses = [
    (ano, mes) for ano in range(2010, 2024)
    for mes in range(1, 13)
    if not (ano == 2023 and mes > 7)
]

limiar_similaridade = 0.55  # Ajust√°vel para mais rigor

# Loop principal m√™s a m√™s
for ano, mes in tqdm(anos_meses, desc="Validando meses com BERT"):
    nome_arquivo = f"tweets_tramontina_{ano}_{mes:02d}.csv.gz"
    caminho_input = os.path.join(input_base, nome_arquivo)
    caminho_output = os.path.join(output_base, nome_arquivo)

    print(f"\nüì¶ Iniciando {ano}/{mes:02d}...", flush=True)

    if not os.path.exists(caminho_input):
        print(f"‚ùå Arquivo n√£o encontrado: {caminho_input}", flush=True)
        continue

    try:
        df = pd.read_csv(caminho_input)
        df = df.dropna(subset=["text"])
        textos = df["text"].astype(str).tolist()
        print(f"üîç {len(textos)} tweets encontrados no m√™s", flush=True)

        # Gera embeddings dos textos
        print("‚öôÔ∏è Gerando embeddings com BERT...", flush=True)
        embeddings_tweets = modelo_bert.encode(textos, convert_to_tensor=True, batch_size=128, show_progress_bar=True)

        # Calcula similaridade com frases-√¢ncora
        print("üìè Calculando similaridades sem√¢nticas...", flush=True)
        similaridades = util.cos_sim(embeddings_tweets, embeddings_ancora).max(dim=1).values
        df["similaridade"] = similaridades.cpu().numpy()

        # Filtra com base no limiar
        df_filtrado = df[df["similaridade"] >= limiar_similaridade]
        print(f"‚úÖ {len(df_filtrado)} tweets relevantes salvos ‚Üí {caminho_output}", flush=True)

        df_filtrado.to_csv(caminho_output, index=False, compression="gzip")

        del df, df_filtrado, embeddings_tweets, similaridades
    except Exception as e:
        print(f"‚ùå Erro ao processar {ano}/{mes:02d} ‚Äî {e}", flush=True)
```

```python
# ‚úÖ Bloco 6 ‚Äî An√°lise de Sentimento com Hugging Face (vers√£o final)
import os
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch
from tqdm import tqdm

# Diret√≥rios de entrada e sa√≠da
input_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/validados/"
output_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/validados_com_sentimento_v4/"
os.makedirs(output_dir, exist_ok=True)

# Detecta GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"üß† Usando dispositivo: {device}")

# Carrega modelo de sentimento da Hugging Face
MODEL_NAME = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME).to(device)
sentiment_fn = pipeline("sentiment-analysis", model=model, tokenizer=tokenizer, device=0 if device == "cuda" else -1)

# Fun√ß√£o robusta para aplicar sentimento
def analisar_sentimento(texto):
    try:
        resultado = sentiment_fn(texto[:512])[0]
        return resultado.get("score", None)
    except:
        return None

# Gera√ß√£o da lista de arquivos (2010 at√© 2023/jul)
anos_meses = [
    (ano, mes) for ano in range(2010, 2024)
    for mes in range(1, 13)
    if not (ano == 2023 and mes > 7)
]

# Loop por arquivo
for ano, mes in tqdm(anos_meses, desc="üîç Analisando sentimento por m√™s"):
    nome_arquivo = f"tweets_tramontina_{ano}_{mes:02d}.csv.gz"
    path_in = os.path.join(input_dir, nome_arquivo)
    path_out = os.path.join(output_dir, nome_arquivo)

    if not os.path.exists(path_in):
        print(f"‚ùå Arquivo n√£o encontrado: {nome_arquivo}")
        continue

    try:
        df = pd.read_csv(path_in)
        if "text" not in df.columns or df.empty:
            print(f"‚ö†Ô∏è {nome_arquivo} ignorado (sem tweets ou sem coluna 'text')")
            continue

        print(f"üì¶ Processando {nome_arquivo} ‚Äî {len(df)} tweets")
        df["score"] = df["text"].astype(str).apply(analisar_sentimento)
        df.to_csv(path_out, index=False, compression="gzip")
        print(f"‚úÖ Sentimentos salvos em: {path_out}\n")

    except Exception as e:
        print(f"‚ö†Ô∏è Erro ao processar {nome_arquivo}: {e}")
```

```python
## ‚úÖ Bloco 7 ‚Äî Mapeamento geogr√°fico por Munic√≠pios (vers√£o atualizada)

import os
import gzip
import json
import requests
import pandas as pd
from tqdm import tqdm
from shapely.geometry import Point, shape

# Diret√≥rios atualizados
input_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/validados_com_sentimento_v4/"
output_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/mapeado_municipios_v1/"
os.makedirs(output_dir, exist_ok=True)

# üîó GeoJSON simplificado dos munic√≠pios do Brasil
url_geojson = "https://raw.githubusercontent.com/dadosfera/brasil-municipios-geojson/refs/heads/main/geojs-100-mun-v2.json"
print("üîÑ Baixando shapefile simplificado dos munic√≠pios...")
geojson_data = requests.get(url_geojson).json()

# ‚õ≥Ô∏è Criar lista de pol√≠gonos com nome de munic√≠pio
municipios = []
for feature in geojson_data["features"]:
    nome = feature["properties"]["name"]
    polygon = shape(feature["geometry"])
    municipios.append((nome, polygon))

# üìÅ Arquivos para processar
arquivos = sorted([f for f in os.listdir(input_dir) if f.endswith(".csv.gz")])
print(f"üì¶ {len(arquivos)} arquivos encontrados para mapeamento geogr√°fico.\n")

# üîÅ Loop de mapeamento
for arq in tqdm(arquivos, desc="Mapeando tweets por munic√≠pio"):
    path_in = os.path.join(input_dir, arq)
    path_out = os.path.join(output_dir, arq)

    try:
        df = pd.read_csv(path_in, compression="gzip")

        if "latitude" not in df.columns or "longitude" not in df.columns:
            print(f"‚ö†Ô∏è Colunas de coordenadas ausentes em: {arq}")
            continue

        # üìå Fun√ß√£o de localiza√ß√£o
        def localizar_municipio(row):
            ponto = Point(row["longitude"], row["latitude"])
            for nome, poligono in municipios:
                if poligono.contains(ponto):
                    return nome
            return "Indefinido"

        df["municipio"] = df.apply(localizar_municipio, axis=1)

        df.to_csv(path_out, index=False, compression="gzip")
        print(f"‚úÖ Mapeado e salvo: {arq}")

    except Exception as e:
        print(f"‚ùå Erro ao processar {arq} ‚Äî {e}")
```

```python
# ‚úÖ NAO RODAR - Bloco 8 ‚Äî Mapeamento por Setor Censit√°rio com multiprocessing

import os
import json
import pandas as pd
from shapely.geometry import Point, shape
from tqdm import tqdm
from multiprocessing import Pool, cpu_count

print(f"üî• N√∫cleos detectados: {cpu_count()}")

# Caminhos
input_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/validados_com_sentimento_v4/"
output_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/mapeado_setores_v2/"
geojson_path = "/n/home03/rpereiraalbuquerque/shapes/municipios_simplificados.geojson"

# Carregar GeoJSON dos setores censit√°rios
print("üì¶ Lendo GeoJSON dos setores censit√°rios...")
with open(geojson_path, 'r', encoding='utf-8') as f:
    setores_data = json.load(f)

# Criar estrutura de setores
setores = {}
for feature in setores_data['features']:
    setor_id = feature['properties'].get('CD_SETOR')
    geometry = shape(feature['geometry'])
    setores[setor_id] = geometry

print(f"‚úÖ Total de setores carregados: {len(setores)}\n")

# Fun√ß√£o auxiliar para um √∫nico tweet
def mapear_um_tweet(tweet):
    try:
        point = Point(tweet["longitude"], tweet["latitude"])
        for setor_id, polygon in setores.items():
            if polygon.contains(point):
                return setor_id
        return None
    except:
        return None

# Fun√ß√£o para processar um √∫nico arquivo
def processar_arquivo(arq):
    path_in = os.path.join(input_dir, arq)
    path_out = os.path.join(output_dir, arq)

    try:
        df = pd.read_csv(path_in)
        if "latitude" not in df.columns or "longitude" not in df.columns:
            print(f"‚ö†Ô∏è Coordenadas ausentes em {arq}")
            return

        pontos = df[["longitude", "latitude"]].to_dict(orient="records")

        with Pool(processes=cpu_count()) as pool:
            setores_mapeados = list(pool.map(mapear_um_tweet, pontos))

        df["CD_SETOR"] = setores_mapeados
        df["mapeado_setor"] = df["CD_SETOR"].notna()
        df.to_csv(path_out, index=False, compression="gzip")

        print(f"‚úÖ Mapeado e salvo: {arq}")
    except Exception as e:
        print(f"‚ùå Erro em {arq}: {e}")

# Criar pasta de sa√≠da
os.makedirs(output_dir, exist_ok=True)

# Rodar todos os arquivos sequencialmente (mas cada um paraleliza internamente)
arquivos = sorted([f for f in os.listdir(input_dir) if f.endswith(".csv.gz")])
print(f"üîç {len(arquivos)} arquivos encontrados para mapeamento por setor.\n")

for arq in tqdm(arquivos, desc="Mapeando tweets por setor"):
    processar_arquivo(arq)
```

```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")

# Caminho dos tweets mapeados por setor censit√°rio
input_dir = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/mapeado_setores_v2/"
output_path = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/gewis_por_setor.csv"

# Lista arquivos .csv.gz
arquivos = sorted([f for f in os.listdir(input_dir) if f.endswith(".csv.gz")])

# Lista para armazenar resultados
resultados = []

print("üìä Calculando GEWI v1, v2 e v3 por setor censit√°rio...\n")

for arquivo in tqdm(arquivos, desc="Processando arquivos"):
    df = pd.read_csv(os.path.join(input_dir, arquivo), compression="gzip")

    if "CD_SETOR" not in df.columns or df.empty:
        continue

    # Cria colunas auxiliares
    df["seguidores_log"] = np.log1p(df["followers"])
    df["ano_mes"] = arquivo.replace("tweets_tramontina_", "").replace(".csv.gz", "")

    # Agrupamento por setor censit√°rio
    for setor, grupo in df.groupby("CD_SETOR"):
        n_total = len(grupo)
        media_sent = grupo["score"].mean()
        log_volume = np.log1p(n_total)
        gewi_v1 = media_sent * log_volume

        # v2 - polaridade positiva e negativa
        positivos = grupo[grupo["score"] >= 0.5]
        negativos = grupo[grupo["score"] < 0.5]

        n_pos = len(positivos)
        n_neg = len(negativos)

        gewi_pos = positivos["score"].mean() * np.log1p(n_pos) if n_pos > 0 else 0
        gewi_neg = negativos["score"].mean() * np.log1p(n_neg) if n_neg > 0 else 0

        # v3 - ponderado por log seguidores
        pesos = grupo["seguidores_log"].fillna(0)
        ponderado = np.average(grupo["score"], weights=pesos) if pesos.sum() > 0 else grupo["score"].mean()
        gewi_v3 = ponderado * log_volume

        resultados.append({
            "ano_mes": grupo["ano_mes"].iloc[0],
            "CD_SETOR": setor,
            "GEWI_v1": gewi_v1,
            "GEWI_v2_pos": gewi_pos,
            "GEWI_v2_neg": gewi_neg,
            "GEWI_v3": gewi_v3,
            "volume_tweets": n_total
        })

# Monta DataFrame final
df_gewis = pd.DataFrame(resultados)
df_gewis.to_csv(output_path, index=False)
print(f"\n‚úÖ √çndices GEWI salvos em: {output_path}")
```

```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm

# Caminho dos arquivos por setor
input_path = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/mapeado_setores_v2/"
output_path = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/"
os.makedirs(output_path, exist_ok=True)

# Lista arquivos
arquivos = sorted([f for f in os.listdir(input_path) if f.endswith(".csv.gz")])

# Lista para armazenar resultados
resultados = []

print("üìä Recalculando GEWI v3 com pesos de influ√™ncia...")

for arq in tqdm(arquivos, desc="Processando arquivos"):
    df = pd.read_csv(os.path.join(input_path, arq))

    # Extrai ano e m√™s do nome do arquivo
    ano_mes = arq.replace("tweets_tramontina_", "").replace(".csv.gz", "")

    if df.empty or "score" not in df.columns or "followers" not in df.columns:
        continue

    # Remove linhas sem localiza√ß√£o v√°lida
    df = df.dropna(subset=["CD_SETOR", "score", "followers"])

    # Aplica o peso de influ√™ncia: log(1 + followers)
    df["peso_influencia"] = np.log1p(df["followers"].clip(lower=0))  # clip evita log negativos

    # Calcula GEWI v3 ponderado por setor
    grupo = df.groupby("CD_SETOR")
    for setor, dados in grupo:
        n = len(dados)
        soma_pesos = dados["peso_influencia"].sum()

        if soma_pesos > 0:
            gewi_v3 = (dados["score"] * dados["peso_influencia"]).sum() / soma_pesos * np.log1p(n)
        else:
            gewi_v3 = 0  # ou np.nan se preferir

        resultados.append({
            "ano_mes": ano_mes,
            "CD_SETOR": setor,
            "GEWI_v3_recalculado": gewi_v3,
            "volume_tweets": n
        })

# Salva novo CSV apenas com o GEWI v3 recalculado
df_final = pd.DataFrame(resultados)
df_final.to_csv(os.path.join(output_path, "gewiv3_recalculado.csv"), index=False)
print("‚úÖ Novo GEWI v3 salvo com sucesso.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caminho para o arquivo com GEWI v3 recalculado
path = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/gewiv3_recalculado.csv"

# Carrega os dados
df = pd.read_csv(path)
df["ano_mes"] = df["ano_mes"].astype(str)

# Ordena cronologicamente
df = df.sort_values("ano_mes")

# Agrupa por m√™s (m√©dia entre setores)
df_mensal = df.groupby("ano_mes").agg({
    "GEWI_v3_recalculado": "mean",
    "volume_tweets": "sum"
}).reset_index()

# üé® Estilo visual
sns.set(style="whitegrid")

# üìà GEWI v3 com pesos
plt.figure(figsize=(14, 5))
plt.plot(df_mensal["ano_mes"], df_mensal["GEWI_v3_recalculado"], marker="o", color="purple")
plt.xticks(rotation=45)
plt.title("‚ö° GEWI v3 ‚Äî Influ√™ncia Real (Followers)")
plt.ylabel("M√©dia GEWI v3 por Setor")
plt.xlabel("ano_mes")
plt.grid(True)
plt.tight_layout()
plt.show()

# üìä Volume de tweets por m√™s (para contexto)
plt.figure(figsize=(14, 4))
plt.plot(df_mensal["ano_mes"], df_mensal["volume_tweets"], marker="s", linestyle="--", color="gray")
plt.title("üìä Volume Total de Tweets por M√™s")
plt.ylabel("N√∫mero de Tweets")
plt.xlabel("ano_mes")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# ‚úÖ Visualiza√ß√£o Final dos GEWIs por m√™s (setores censit√°rios)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Caminho do arquivo final com GEWIs
caminho = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/gewis_por_setor.csv"

# Leitura da base
df = pd.read_csv(caminho)

# üì¶ Ajuste de nome das colunas para consist√™ncia
df = df.rename(columns={
    "GEWI_v1": "gewiv1",
    "GEWI_v2_pos": "gewiv2_pos",
    "GEWI_v2_neg": "gewiv2_neg",
    "GEWI_v3": "gewiv3",
    "volume_tweets": "n_tweets"
})

# üìä Agrega√ß√£o por m√™s
df_mensal = df.groupby("ano_mes").agg({
    "n_tweets": "sum",
    "gewiv1": "mean",
    "gewiv2_pos": "mean",
    "gewiv2_neg": "mean",
    "gewiv3": "mean"
}).reset_index()

# üé® Estilo
sns.set(style="whitegrid")

# üìà Volume total de tweets
plt.figure(figsize=(14, 5))
sns.lineplot(data=df_mensal, x="ano_mes", y="n_tweets", marker="o", linewidth=2)
plt.xticks(rotation=45)
plt.title("üìä Volume total de tweets validados por m√™s")
plt.ylabel("Total de Tweets")
plt.tight_layout()
plt.grid(True)
plt.show()

# üìê GEWI v1
plt.figure(figsize=(14, 5))
sns.lineplot(data=df_mensal, x="ano_mes", y="gewiv1", marker="o", color="blue")
plt.xticks(rotation=45)
plt.title("üìà GEWI v1 ‚Äî Score Composto (sentimento √ó volume)")
plt.ylabel("GEWI v1 (m√©dia)")
plt.tight_layout()
plt.grid(True)
plt.show()

# üí† GEWI v2 ‚Äî Positivo e Negativo
plt.figure(figsize=(14, 5))
sns.lineplot(data=df_mensal, x="ano_mes", y="gewiv2_pos", label="Positivo", color="green", marker="o")
sns.lineplot(data=df_mensal, x="ano_mes", y="gewiv2_neg", label="Negativo", color="red", marker="o")
plt.xticks(rotation=45)
plt.title("üìà GEWI v2 ‚Äî Decomposi√ß√£o Positiva vs Negativa")
plt.ylabel("GEWI v2 (m√©dia)")
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.show()

# üß† GEWI v3 ‚Äî Ponderado por Influ√™ncia
plt.figure(figsize=(14, 5))
sns.lineplot(data=df_mensal, x="ano_mes", y="gewiv3", marker="s", color="purple")
plt.xticks(rotation=45)
plt.title("üí• GEWI v3 ‚Äî Influ√™ncia Ponderada (log(1 + seguidores))")
plt.ylabel("GEWI v3 (m√©dia)")
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
## Verifica√ß√£o da Pondera√ß√£o por Influ√™ncia no GEWI v3
```

```python
import pandas as pd
import numpy as np

# Carrega um m√™s com dados mapeados e com GEWI calculado
path = "/n/netscratch/cga/Lab/rpalbuquerque/GEWI 2.0/mapeado_setores_v2/tweets_tramontina_2020_01.csv.gz"
df = pd.read_csv(path)

# Cria a coluna de peso usado no GEWI v3
df["peso"] = np.log1p(df["followers"])

# Exibe estat√≠sticas b√°sicas dos pesos
print(df["peso"].describe())

# Compara√ß√£o entre m√©dia simples e ponderada
sentimento_medio = df["score"].mean()
sentimento_ponderado = np.average(df["score"], weights=df["peso"])

print(f"\nüìå Sentimento m√©dio simples: {sentimento_medio:.4f}")
print(f"üìå Sentimento m√©dio ponderado (GEWI v3): {sentimento_ponderado:.4f}")
```
